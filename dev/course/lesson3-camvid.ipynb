## Datasets

```python
size = src_size//2
bs=8
```

```python
valid_fnames = (path/'valid.txt').read().split('\n')
```

```python
# export
def FileSplitter(fname):
    "Split `items` depending on the value of `mask`."
    valid = Path(fname).read().split('\n') 
    def _func(x): return x.name in valid
    def _inner(o, **kwargs): return FuncSplitter(_func)(o)
    return _inner
```

```python
camvid = DataBlock(blocks=(ImageBlock, ImageBlock(cls=PILMask)),
                   get_items=get_image_files,
                   splitter=FileSplitter(path/'valid.txt'),
                   get_y=lambda o: path/'labels'/f'{o.stem}_P{o.suffix}')
```

```python
dbunch = camvid.databunch(path/"images", bs=8, path=path,
                          batch_tfms=[*aug_transforms(size=(360,480)), Normalize(*imagenet_stats)])
```

## Model

```python
name2id = {v:k for k,v in enumerate(codes)}
void_code = name2id['Void']

def acc_camvid(input, target):
    target = target.squeeze(1)
    mask = target != void_code
    return (input.argmax(dim=1)[mask]==target[mask]).float().mean()
```

```python
metrics=acc_camvid
# metrics=accuracy
```

```python
wd=1e-2
```

```python
dbunch.vocab = codes
```

```python
learn = unet_learner(dbunch, resnet34, metrics=metrics)
```

```python
learn.lr_find()
```

```python
lr=3e-3
```

```python
learn.fit_one_cycle(10, slice(lr), pct_start=0.9, wd=wd)
```

```python
learn.save('stage-1')
```

```python
learn.load('stage-1');
```

```python
learn.show_results(max_n=4, figsize=(9,4))
```

```python
learn.unfreeze()
```

```python
lrs = slice(lr/400,lr/4)
```

```python
learn.fit_one_cycle(12, lrs, pct_start=0.8, wd=wd)
```

```python
learn.save('stage-2');
```
